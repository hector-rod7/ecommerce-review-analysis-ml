En este notebook se realiza el preprocesamiento de las opiniones de los clientes con el objetivo de preparar los textos para las fases posteriores de análisis y modelado.

Se aplicarán diferentes técnicas de limpieza para reducir el ruido presente en las reviews y facilitar que los algoritmos de Machine Learning puedan extraer información relevante.


```python
import pandas as pd
import re
import nltk
```

```python
df = pd.read_csv("amazon_reviews_eda.csv")

df.head()
```

```python
df.info()
print(df.loc[0, "review"])
```

```python
#limpieza básica
def limpiar_texto_basico(texto):
    texto = texto.lower()
    texto = texto.replace("\n", " ")
    texto = texto.replace("\r", " ")
    texto = re.sub(r"\s+", " ", texto)
    texto = texto.strip()
    return texto
```

```python
df["review_clean_basic"] = df["review"].apply(limpiar_texto_basico)

df[["review", "review_clean_basic"]].head()
```

```python
#comparar visualización del texto original con la limpieza
print("REVIEW ORIGINAL:")
print(df.loc[0, "review"])

print("\nREVIEW LIMPIA:")
print(df.loc[0, "review_clean_basic"])
```

```python
#eliminar signos de puntuación
def eliminar_puntuacion(texto):
    texto = re.sub(r"[^\w\s]", "", texto)
    return texto
```

```python
df["review_no_punctuation"] = df["review_clean_basic"].apply(eliminar_puntuacion)
```

```python
print("ANTES")
print(df.loc[0, "review_clean_basic"])

print("\nDESPUÉS")
print(df.loc[0, "review_no_punctuation"])
```

```python
#eliminar numeros
def eliminar_numeros(texto):
    texto = re.sub(r"\d+", "", texto)
    return texto

df["review_no_numbers"] = df["review_no_punctuation"].apply(eliminar_numeros)
```

#aplicar stopwords


```python
#stopwords
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
```

```python
#crear función para eliminarlas
def eliminar_stopwords(texto):
    palabras = texto.split()
    palabras_limpias = []

    for palabra in palabras:
        if palabra not in stop_words:
            palabras_limpias.append(palabra)

    texto_limpio = " ".join(palabras_limpias)
    return texto_limpio
```

```python
#aplicar función
df["review_no_stopwords"] = df["review_no_numbers"].apply(eliminar_stopwords)
```

```python
print("ANTES")
print(df.loc[0, "review_no_numbers"])

print("\nDESPUÉS")
print(df.loc[0, "review_no_stopwords"])
```

Llevaré a cabo la lematización

```python
import nltk

nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
def lematizar_texto(texto):

    palabras = texto.split()

    palabras_lematizadas = []

    for palabra in palabras:
        palabras_lematizadas.append(lemmatizer.lemmatize(palabra))

    texto_final = " ".join(palabras_lematizadas)

    return texto_final
```

```python
df["review_clean"] = df["review_no_stopwords"].apply(lematizar_texto)
```

```python
print("ORIGINAL")
print(df.loc[0, "review"])

print("\nLIMPIA")
print(df.loc[0, "review_clean"])
```

Elimino columnas ya que ya no sirven más

```python
columnas_eliminar = [
    "review_clean_basic",
    "review_no_punctuation",
    "review_no_numbers",
    "review_no_stopwords"
]

df.drop(columns=columnas_eliminar, inplace=True)
```

```python
# Crear la variable sentiment

def clasificar_sentimiento(rating):

    if rating <= 2:
        return "Negativo"

    elif rating == 3:
        return "Neutro"

    else:
        return "Positivo"

df["sentiment"] = df["rating"].apply(clasificar_sentimiento)
```
